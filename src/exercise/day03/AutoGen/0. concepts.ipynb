```python
# sudo ln -sf /usr/bin/python3 /usr/bin/python

import os
import autogen
from autogen import AssistantAgent, UserProxyAgent

llm_config = { "config_list": [{ "model": os.environ.get("OPENAI_DEFAULT_MODEL"), "api_key": os.environ.get("OPENAI_API_KEY") }] }
assistant = AssistantAgent("assistant", llm_config=llm_config)

user_proxy = UserProxyAgent( 
    "user_proxy", code_execution_config={"executor": autogen.coding.LocalCommandLineCodeExecutor(work_dir="coding")},
    human_input_mode="NEVER"
) 

# Start the chat
user_proxy.initiate_chat(
    assistant,
    message="엔비디아와 테슬라의 연간 주가를 그래프로 그려줘.",
) # 사용자는 실행결과를 수동으로 입력하거나, 또는 Auto reply로 코드블록을 실행하고 자동으로 결과가 입력되게 할수 있음
```

```python
# root로 docker 설치 (https://get.docker.com)
# sudo groupadd docker (있으면 불필요)
# sudo usermod -aG docker $USER
# newgrp docker
# sudo chmod 666 /var/run/docker.sock

import os
import autogen
from autogen import AssistantAgent, UserProxyAgent

llm_config = { "config_list": [{ "model": os.environ.get("OPENAI_DEFAULT_MODEL"), "api_key": os.environ.get("OPENAI_API_KEY") }] }

with autogen.coding.DockerCommandLineCodeExecutor(work_dir="coding") as code_executor:
    assistant = AssistantAgent("assistant", llm_config=llm_config)
    user_proxy = UserProxyAgent(
        "user_proxy", code_execution_config={"executor": code_executor},
        human_input_mode="NEVER"
    )

    # Start the chat
    user_proxy.initiate_chat(
        assistant,
        message="엔비디아와 테슬라의 연간 주가를 그래프로 그려줘. 그래프를 plot.png이라는 파일로 저장해줘.",
    )
```

```python
from agents import Agent, ImageGenerationTool, Runner
from IPython.display import Image, display
import base64

agent = Agent(
    name="Image Styler",
    model="gpt-4o-mini",
    instructions="You are a helpful agent that restyles input images.",
    tools=[
        ImageGenerationTool(
            tool_config={"model":"gpt-image-1-mini", "type": "image_generation", "quality": "medium"},
        )
    ],
)

prompt = "이 이미지를 수채화 스타일로 바꿔줘"

with open("input_image.jpg", "rb") as f:
    image_bytes = f.read()
    image_base64 = base64.b64encode(image_bytes).decode("utf-8")

input_data = [
    {
        "role": "user",
        "content": [
            {"type": "input_text", "text": prompt},
            {"type": "input_image", "image_url": f"data:image/jpeg;base64,{image_base64}"},
        ],
    }
]
result = await Runner.run(agent,
    input=input_data
)
```

```python
from IPython.display import Image, display
import base64

for item in result.new_items:
    if (
        item.type == "tool_call_item"
        and item.raw_item.type == "image_generation_call"
        and (img_result := item.raw_item.result)
    ):
        image_data = base64.b64decode(img_result)
        display(Image(image_data))
```

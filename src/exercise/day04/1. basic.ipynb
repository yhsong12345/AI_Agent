```python
from agents import GuardrailFunctionOutput, Agent, Runner
from pydantic import BaseModel


class HistoryMathOutput(BaseModel):
    is_history_math: bool
    reasoning: str

guardrail_agent = Agent(
    name="Guardrail check",
    instructions=(
        "당신은 사용자의 질문이 역사나 수학에 대한 것인지 판별하는 감시자입니다.\n"
        "반드시 다음 JSON 형식으로만 응답하세요:\n"
        "{ \"is_history_math\": true 또는 false, \"reasoning\": \"판단 이유\" }"
    ),
    output_type=HistoryMathOutput,
)

async def history_math_guardrail(ctx, agent, input_data):
    result = await Runner.run(guardrail_agent, input_data, context=ctx.context)
    final_output = result.final_output_as(HistoryMathOutput)
    return GuardrailFunctionOutput(
        output_info=final_output,
        tripwire_triggered=not final_output.is_history_math
    )
```

```python
from agents import Agent, InputGuardrail, GuardrailFunctionOutput, Runner
from agents.exceptions import InputGuardrailTripwireTriggered
from pydantic import BaseModel
import asyncio

triage_agent = Agent(
    name="triage_agent",
    instructions="사용자의 숙제 질문에 따라 어떤 에이전트를 사용할 것인지 결정합니다",
    handoffs=[history_tutor_agent, math_tutor_agent],
    input_guardrails=[
        InputGuardrail(guardrail_function=history_math_guardrail),
    ],
)
```

```python
try:
    result = await Runner.run(triage_agent, "대한민국의 첫번째 대통령은 누구인가?")
    print(result.final_output)
except InputGuardrailTripwireTriggered as e:
    print("Guardrail blocked this input:", e)
```

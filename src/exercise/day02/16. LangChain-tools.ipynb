```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model="gpt-4o-mini",
    temperature=0,
)
```

```python
from langchain_core.tools import tool

@tool
def add_numbers(a: int, b: int) -> int:
    """두개의 숫자를 더한다."""
    return a + b
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model=llm,
    tools=[add_numbers],
    system_prompt="당신은 친절한 어시스턴트입니다. 사용자의 질문에 친절하게 답변하세요."
)

response = agent.invoke({
    "messages": [{"role": "user", "content": "3 + 4는 얼마?"}]
})
print(response["messages"][-1].content)
```

```python
for message in response["messages"]:
    print(message)
```

```python
import json

@tool
def get_current_weather(location, unit="섭씨"):
    """주어진 장소의 날씨 정보를 제공합니다."""
    weather_info = {
        "장소": location,
        "온도": "24",
        "단위": unit,
        "예보": ["비"],
    }
    return json.dumps(weather_info, ensure_ascii=False)
```

```python
from langchain.agents import create_agent

agent = create_agent(
    model=llm,
    tools=[get_current_weather],
    system_prompt="당신은 친절한 어시스턴트입니다. 사용자의 질문에 친절하게 답변하세요."
)

response = agent.invoke({
    "messages": [{"role": "user", "content": "서울의 날씨를 알려줘"}]
})
print(response["messages"][-1].content)
```

```python
gen = agent.stream({
    "messages": [{"role": "user", "content": "서울의 날씨를 알려줘"}]
})

for i, message in enumerate(gen):
    print(f"{i}: {message}")
```
